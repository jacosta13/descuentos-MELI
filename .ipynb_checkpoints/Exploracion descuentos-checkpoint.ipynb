# Evaluacion de variables 

### Enlaces

```python
cols_enlaces = ["permalink", "thumbnail"]

df.drop(
    columns=cols_enlaces,
    inplace=True
)

cols_enlaces
```

### Titulo

```python
df["title"]
```

```python
def num_words(x:str):
    return len(x.split(" "))
```

```python
num_words("Bomba De Succión Al Vacío Para Baños")
```

```python
df["num_words_tittle"] = df["title"].map(num_words)

df["num_words_tittle"]
```

```python
def num_ints(x:str):
    return sum(c.isdigit() for c in x)
```

```python
num_ints("Bo87mba De Succ5ión Al Va5cío Par67a Baños")
```

```python
df["num_ints_tittle"] = df["title"].map(num_ints)
df["num_ints_tittle"]
```

```python
def num_letters(x:str):
    return sum(c.isalpha() for c in x)
```

```python
num_letters("Bo87mba De Succ5ión Al Va5cío Par67a Baños")
```

```python
df["num_letters_tittle"] = df["title"].map(num_letters)
df["num_ints_tittle"]
```

## Separacion categoricas, numericas y var respuesta

```python
df["shipping_free_shipping"] = df["shipping_free_shipping"].map(
    lambda x : 1 if x else 0
)
```

```python
df.columns
```

```python
df.dtypes
```

```python
cols_categoricas = [
    "title",
    "available_quantity",
    "sold_quantity",
    "category_id",
    "address_state_name",
    "address_city_name",   
    "shipping_mode",
    "shipping_logistic_type",
    "seller_address_state_name"
]


cols_numericas = [
    "price",
    "original_price",
    "order_backend",
    "installments_quantity",
    "installments_amount",
    "shipping_free_shipping",
    "NUM_ATTRIBUTEs",
    "num_tags",
    "best_seller_candidate",
    "loyalty_discount_eligible",
    "good_quality_picture",
    "MODEL_bool",
    "BRAND_bool",
    "num_words_tittle",
    "num_ints_tittle",
    "num_letters_tittle"
]
```

```python
for col in cols_categoricas:
    df[col] = df[col].astype(object)
```

```python
df[cols_categoricas].info()
```

```python
for col in cols_numericas:
    df[col] = df[col].astype(int)
```

```python
df[cols_numericas].info()
```

```python
df_gp_cat = df[
    ["category_id","perc_disc"]
].groupby(
    by=["category_id"]
).agg(
    ["mean"]
).reset_index(
    drop=False
)

df_gp_cat.columns = ["category_id", "mean_perc_disc"]

df_gp_cat
```

```python
df_disc_cat = pd.merge(
    left=df_gp_cat,
    left_on="category_id",
    right=df_cat[["cod_category", "name_category"]],
    right_on="cod_category"
)

df_disc_cat.sort_values(
    by=["mean_perc_disc"],
    ascending=False
)
```

```python
df_disc_cat.sort_values(
    by=["mean_perc_disc"],
    ascending=True
)
```
